### 變數的應用
Q：邊長為3,4,5的三角形，其面積大小為何?
<br>海龍公式：
<br>s=(a+b+c)/2
<br>sqrt(s(s-a)(s-b)(s-c))

```python
import math
a,b,c=3,4,5
s=(a+b+c)/2
d=math.sqrt(s*(s-a)*(s-b)*(s-c))
print(d)
```

Q：半徑為10的圓形，其面積大小為何?

```python
import math
r=20
print(r*r*math.pi)
```

## 轉換型態

```python
#小心不同型態之間所得到的結果跟想像不同
x="must"
print(10*x)
print("10"+x)
print(10,x)
```

```python
x=12
print(type(x))
y=float(x)+0.5 #強制型別轉換
print(type(y))
print(int("123"))
print(int(123.33))
print(int(float("123.33")))
```

```python
#字串的型別轉換
x=10*3.25  #x=32.5
y=200*200  #y=40000
s="The value of x is " + str(x) + ' and y is ' + str(y)
print(s)
print("The value of x is", x,"and y is", y)
print("The value of x is {} and y is {}".format(x,y))
```
