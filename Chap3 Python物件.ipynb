## list的長度

```python
my_list2 = ['python','javascript','C#']
print(len(my_list2)) #len=返回對象（字符、列表、元組等）長度或項目個數。
print(len(my_list2[0]))
```

## 附加

```python
my_list2 = ['python','javascript','C#']
my_list2.append('java') #append=在列表末尾增加一個資料項
print(my_list2)
```

```python
my_list2 = ['python','javascript','C#']
my_list2.insert(1,'java') #insert=在指定對象插入列表的指定位置
print(my_list2)
```

## 刪除

```python
my_list2 = ['python','javascript','C#']
my_list2.remove('python') #刪除指定元素
print(my_list2)
del my_list2[-1] #刪除指定位置的元素
print(my_list2)
my_list2.clear() #清除全部
print(my_list2)
```

## list取值

```python
a = [1,2,3,4,5,6,7,8,9]
b = a[0:3] #範圍取值，0,1,2不包含3
print(b)
c = a[0:9:2] #間隔2取值
print(c)
d = a[0:] #0到最後
print(d)
e = a[:9] #從頭到9
print(e)
f = a[:]  #取全部值
print(f)
```

```python
a = [1,2,3,4,5,6,7,8,9]
del a[7:9] #刪除某範圍值7-9
print(a)
print(min(a)) #最小值
print(max(a)) #最大值
print(a.index(1)) #找出1所在的第一個索引值
b= [1,0,0,1,1,1]
print(b.count(1)) #1出現的次數
a.reverse() #串列反轉
print(a)
a.sort() #串列由小到大排序，a的值會被改變
print(a)
```

```python
expenses = [1138,101,12,300,2049]
s = sorted(expenses) #sorted=對元素排序之後，傳回一個排序好的新list，而原本的list則不受影響
print(s,expenses)
```

## Tuple(元祖)
tuple(元祖)可以視為不可改變的串列(list)。tuple跟list很像，但是不能新增刪除或者更新tuple的元素。

tuple比起list的好處：
* 佔用空間少
* 可以當作dictionary(字典)的key
* 具名tuple當作物件替代
* 當作函式的引數

```python
a_list=['python','js','C#']
a_tuple=tuple(a_list)
b_tuple=('python','js','C#')
print(a_tuple)
print(b_tuple)
print(a_tuple[0])
```

## Dictionary - dict(字典)
dict為帶有鍵值(key)的串列
* 表示方法為{}

```python
languages={}
languages={'name':'python','version':'3.7'}
a_list=['python','3.7']
a_dict={0:'python',1:'3.7'}
print(a_list[0])
print(a_list[1])
print(a_dict)
print(a_dict[0]) #dict就是索引值可以改變的list
```

```python
languages={'name':'python','version':'3.7'} #鍵(key)用:分割，值(value)用,分割
print(languages['name'])
print(languages)
print(languages.keys())  #key=以列表返回一個字典所有的鍵
print(languages.values()) #values=以列表返回字典中的所有值
print(languages.items()) #items=以列表返回可遍歷的(鍵,值)元組數組
print('name' in languages)
```

```python
import random
dice=random.randint(1,6)
print("骰出骰子點數：")
if dice==1:
    print('一點')
elif dice==2:
    print('二點')
elif dice==3:
    print('三點')
elif dice==4:
    print('四點')
elif dice==5:
    print('五點')
else:
    print('六點')
```
